```python
# Dependencies and Setup
%matplotlib inline
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
```

```python
# Hide warning messages in notebook
import warnings
warnings.filterwarnings('ignore')
```

```python
# File to Load (Remember to Change These)
mouse_drug_data_to_load = "C:\\Users\\Portales Ely\\Desktop\\mouse_drug_data.csv"
clinical_trial_data_to_load = "C:\\Users\\Portales Ely\\Desktop\\clinicaltrial_data.csv"
```

```python
# Read the Mouse and Drug Data and the Clinical Trial Data

mouse_drug_data_to_load_df = pd.read_csv(mouse_drug_data_to_load)
```

```python
mouse_drug_data_to_load_df.head()
```

```python
clinical_trial_data_to_load_df = pd.read_csv(clinical_trial_data_to_load)
```

```python
clinical_trial_data_to_load_df.head()
```

```python
# Combine the data into a single dataset

merge_table = pd.merge(clinical_trial_data_to_load_df,mouse_drug_data_to_load_df, on="Mouse ID")
merge_table


# Display the data table for preview
```

```python
# Tumor Response to Treatment

# Store the Mean Tumor Volume Data Grouped by Drug and Timepoint
# Convert to DataFrame
# Preview DataFrame

group_drug=merge_table.groupby(["Drug", "Timepoint"])
tumor_siz=group_drug["Tumor Volume (mm3)"]
mean_tumor=pd.DataFrame({"Mean Tumor Volume (mm3)":tumor_siz.mean()}).reset_index()
mean_tumor
```

```python
# Store the Standard Error of Tumor Volumes Grouped by Drug and Timepoint
# Convert to DataFrame
# Preview DataFrame

error_tumor_size=pd.DataFrame({"SEM Tumor Volume (mm3)":tumor_siz.sem()}).reset_index()
error_tumor_size.head()
```

```python
# Minor Data Munging to Re-Format the Data Frames
# Preview that Reformatting worked

minor_data=mean_tumor.pivot(index="Timepoint",columns="Drug", values="Mean Tumor Volume (mm3)")
sem_minor_data=error_tumor_size.pivot(index="Timepoint",columns="Drug", values="SEM Tumor Volume (mm3)")
minor_data.head()
```

```python
# Generate the Plot (with Error Bars)
# Save the Figure

x_axis=list(minor_data.index)
capomulin_tumor_res=plt.errorbar(x_axis, minor_data["Capomulin"], yerr=sem_minor_data["Capomulin"], linewidth=0.5, marker='o',color='red',label="Capomulin")
infubinol_tumor_res=plt.errorbar(x_axis, minor_data["Infubinol"], yerr=sem_minor_data["Infubinol"], linewidth=0.5, marker='^',color='blue',label="Infubinol")
ketapril_tumor_res=plt.errorbar(x_axis, minor_data["Ketapril"], yerr=sem_minor_data["Ketapril"], linewidth=0.5, marker='s', color='green', label="Ketapril")
placebo_tumor_res=plt.errorbar(x_axis, minor_data["Placebo"], yerr=sem_minor_data["Placebo"], linewidth=0.5, marker='d', color='black', label="Placebo")

plt.ylim(33, 73)
plt.grid()
plt.legend(loc="upper left")
plt.title("Tumor Response by Treatment")
plt.xlabel("Time (Days)")
plt.ylabel("Tumor Volume (mm3)")

# Save the Figure
plt.savefig("C:\\Users\\Portales Ely\\Desktop\\tumor_res.png")
```

```python
# Store the Mean Met. Site Data Grouped by Drug and Timepoint 
# Convert to DataFrame
# Preview DataFrame


met_data=group_drug["Metastatic Sites"]
mean_met_data=pd.DataFrame({"Mean Metastatic Sites":met_data.mean()}).reset_index()
pd.DataFrame(met_data.mean()).head()

```

```python
# Store the Standard Error associated with Met. Sites Grouped by Drug and Timepoint 
# Convert to DataFrame
# Preview DataFrame

sem_data=pd.DataFrame({"SEM Metastatic Sites":met_data.sem()}).reset_index()
pd.DataFrame(met_data.sem()).head()
```

```python
# Minor Data Munging to Re-Format the Data Frames
# Preview that Reformatting worked

minor_data_drug=mean_met_data.pivot(index="Timepoint",columns="Drug", values="Mean Metastatic Sites")
sem_dat_drug=sem_data.pivot(index="Timepoint",columns="Drug", values="SEM Metastatic Sites")

minor_data_drug.head()

```

```python
# Generate the Plot (with Error Bars)
# Save the Figure
# Show the Figure


capomulin_met=plt.errorbar(x_axis, minor_data_drug["Capomulin"], yerr=sem_dat_drug["Capomulin"], linewidth=0.5, marker='o',color='red',label="Capomulin")
infubinol_met=plt.errorbar(x_axis, minor_data_drug["Infubinol"], yerr=sem_dat_drug["Infubinol"], linewidth=0.5, marker='^',color='blue',label="Infubinol")
ketapril_met=plt.errorbar(x_axis, minor_data_drug["Ketapril"], yerr=sem_dat_drug["Ketapril"], linewidth=0.5, marker='s', color='green', label="Ketapril")
placebo_met=plt.errorbar(x_axis, minor_data_drug["Placebo"], yerr=sem_dat_drug["Placebo"], linewidth=0.5, marker='d', color='black', label="Placebo")

plt.grid()
plt.legend(loc="upper left")
plt.title("Metastatic Sites by Treatment")
plt.xlabel("Time (Days)")
plt.ylabel("Metastatic Sites")

# Save the Figure
plt.savefig("C:\\Users\\Portales Ely\\Desktop\\tumors_resp.png")
```

```python
# Store the Count of Mice Grouped by Drug and Timepoint (W can pass any metric)
# Convert to DataFrame
# Preview DataFrame

mice_count=group_drug["Mouse ID"]
mice_count=pd.DataFrame({"Mouse Count":mice_count.count()}).reset_index()
mice_count.head()
```

```python
# Calculate the percent changes for each drug
# Display the data to confirm

changes=minor_data.loc[45,:]/minor_data.loc[0,:]*100-100
changes

```

```python

# Store all Relevant Percent Changes into a Tuple
tumors_changes=(changes["Capomulin"],changes["Infubinol"],changes["Ketapril"],changes["Placebo"])

# Splice the data between passing and failing drugs
x_axis=np.arange(4)
x_labels=["Capomulin", "Infubinol", "Ketapril", "Placebo"]
tick_locations=[value for value in x_axis]
plt.bar(x_axis, tumors_changes, color=['green' if tumors_changes[x]<0 else 'red' for x in x_axis])


# Orient widths. Add labels, tick marks, etc. 
plt.axhline(y=0, color='black')
plt.title("Tumor Change Over 45 Day Treatment")
plt.ylabel("% Tumor Volume Change")
plt.xticks(tick_locations, x_labels)
plt.ylim(-30, 70)


# Use functions to label the percentages of changes
def autolabel(values_list):
    for idx, value in enumerate(values_list):
        if value > 0:
            plt.text(idx-0.25, 5, f'{round(value,2)}%', color='white', fontsize=12, fontweight='bold')
        else:
            plt.text(idx-0.35, -10, f'{round(value,2)}%', color='white', fontsize=12, fontweight='bold')


# Call functions to implement the function calls
autolabel(tumors_changes)

# Save the Figure
plt.savefig("C:\\Users\\Portales Ely\\Desktop\\change_tumors")

# Show the Figure
plt.show()
```
